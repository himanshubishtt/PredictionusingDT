```python
# Importing necessary libraries
import sklearn.datasets as datasets
import pandas as pd
import matplotlib.pyplot as plt
```

# Loading and observing the dataset

```python
# Loading the iris dataset
iris=datasets.load_iris()

# Forming the input dataframe
df=pd.DataFrame(iris.data, columns=iris.feature_names)
df.head(10)
```

```python
# Forming the output labels array
y=iris.target
y
```

# Importing the Decision Tree classifier and training it on the dataset
The classifier is going to construct a tree where each node represent a test on the input value on a feature which decides the output label of the input given

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
dtc.fit(df,y)
print("Classifier Created")
```

# Visualizing the tree

```python
from sklearn import tree
plt.figure(figsize = (20,20))
tree.plot_tree(dtc , filled = True)
plt.title("Diagramatic representation of Decision Tree for the given data")
```

# Now, we can feed new input/test data to this classifier and it will determine the output label accordingly
